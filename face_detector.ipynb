```python
from google.colab.patches import cv2_imshow
import cv2

img = cv2.imread('/content/Cahyo Gadhang Putro Baskoro.jpg')

if img is None:
    print("Error: Image not loaded. Please check the file path.")
else:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print("Image converted to grayscale successfully.")

    haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    faces_rect = haar_cascade.detectMultiScale(gray, 1.1, 9)

    for (x, y, w, h) in faces_rect:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Resize image with fixed width (600px) while keeping aspect ratio
    fixed_width = 600
    aspect_ratio = fixed_width / img.shape[1]
    new_height = int(img.shape[0] * aspect_ratio)

    img_resized = cv2.resize(img, (fixed_width, new_height), interpolation=cv2.INTER_AREA)

    cv2_imshow(img_resized)
```
